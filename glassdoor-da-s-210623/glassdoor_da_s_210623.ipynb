## Task 1: Removing number from the company name

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split("\n")[0])
df.head()
```

```python
df['Company Name']
```

## Task 2: Transforming the salary estimates column into int


```python
str(df['Salary Estimate'])
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df['Salary Estimate']
```

```python
df.head()
df = df.copy()
```

```python
df['Min Sal'] = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')[0].astype(int)
df['Max Sal'] = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')[1].astype(int)
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K',''))
# better option
# df['Salary Estimate'].apply(lambda x: x.split('(')[0]).unique()
# df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
# df['Salary Estimate'].apply(lambda x: x.replace('K','').replace('$','')).unique()
# df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('K','').replace('$',''))
# df['min_salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
# df['max_salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
df.head()
```

## Task 3: City and State column from Location

```python
df['loc_city'] = df['Location'].apply(lambda x: x.split(',')[0])
df['loc_state'] = df['Location'].apply(lambda x: x.split(',')[-1])
df.drop(columns = 'Location')
df.head()
```

## Task 4: Company Age

```python
df['company_age'] = abs(df['Founded']-2023) # could just swapped but had to use 'abs'
df.head()
```

## Task 5: Extraction from Job Description

```python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['big_data'] = df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df.head()
```
