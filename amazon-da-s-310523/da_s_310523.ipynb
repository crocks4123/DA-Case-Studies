### Task 1: Merge the 12 months of sales data into a single CSV file 

```python
# Importing the files from a directory using "os"
files = [file for file in os.listdir('./Sales_analysis')]

all_months_data = pd.DataFrame() # creating an empty dataframe to store all csv together

# for file in files:
#   print(file)

for file in files:
  df = pd.read_csv('./Sales_analysis/'+file)
  all_months_data = pd.concat([all_months_data, df]) # concatinating the existing and looping csvs
all_months_data.head()


all_months_data.to_csv('alinvan.csv', index=False) # to csv and removing the default index column
```

## Q2: Which city had most sales?

***adding a column with only city names***

```python
# .apply function using lambda to access elements from that row

def city_name(add):
  return add.split(',')[1]

def state_name(add):
  return add.split(',')[2].split(' ')[1] # only city and state not the pincode
# all_data2['City'] = all_data2['Purchase Address'].apply(lambda x: city_name(x)+ ' (' + state_name(x) + ')').  <-- old method
all_data2['City'] = all_data2['Purchase Address'].apply(lambda x: f"{city_name(x)} ({state_name(x)})")


# all_data2['City'] = all_data2['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data2.head()
```

```python
results2 = all_data2.groupby('City').sum()
results2
```

Plotting graph to provide insights

```python
import matplotlib.pyplot as plt
city = [city for city, df in all_data2.groupby('City')]
plt.bar(city, results2['Sales'])
plt.xticks(city, rotation='vertical', size=8) # best to play with the labels
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.show()
```

## Q4: What products are often sold together?

Note that in one order there are multiple products. Hence we have to access the unique order ids and there respective PRODUCT

```python
df = all_data2[all_data2['Order ID'].duplicated(keep = False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
# from stackoverflow

from itertools import combinations
from collections import Counter
count = Counter()

for row in df['Grouped']:
  row_list = row.split(' ,')
  count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
  print(key, value)
```

***Overlaying the price graph on this to check why AAA is the most sold item***

```python
prices = all_data2.groupby('Product').mean()['Price Each'] # this is for the avg price of that item
prices

# stackoverflow for the subplot

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(sal, qorder, color = 'r')
ax2.plot(sal, prices, 'b-')

ax1.set_xlabel('Products', color = 'g')    
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Prices', color = 'g')
ax1.grid()
ax1.set_xticklabels(sal, rotation = 'vertical',  size = 8)

plt.show()
```
